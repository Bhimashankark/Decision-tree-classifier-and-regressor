```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
import seaborn as sns
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
X=df.iloc[:,:-1]
y=dataset.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier()
```

```python
classifier.fit(X_train,y_train)
```

```python
X_train.head()
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier,filled=True)
```

```python
classifier=DecisionTreeClassifier(max_depth=2,criterion='entropy')
classifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,11))
tree.plot_tree(classifier,filled=True)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
y_pred.shape
```

## DecisionTree Prepruning And Hyperparameter Tuning For Huge Data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
parameters={
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5,6],
    'max_features':['auto', 'sqrt', 'log2']
  
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf=GridSearchCV(classifier,param_grid=parameters,cv=2,scoring='accuracy')
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred=clf.predict(X_test)
```

```python
classifier=DecisionTreeClassifier(criterion='gini',max_depth=6,max_features='log2',splitter='random')
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
